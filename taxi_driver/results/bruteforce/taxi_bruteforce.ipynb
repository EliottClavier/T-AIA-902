```python
from pathlib import Path

from common.params import TaxiDriverParams
from common.algorithms import BruteForce
from common.policies import Random
from common.environments import TaxiDriver
from common.plots import TaxiDriverPlots
```

```python
params = TaxiDriverParams(
    n_episodes=10,
    n_runs=1,
    learning_rate=0.85,
    gamma=0.99,
    epsilon=0.1,
    min_epsilon=0.001,
    random_seed=True,
    seed=123,
    max_n_steps=20000,
    savefig_folder=Path("./static/img/taxi_driver/"),
    savemodel_folder=Path("./static/models/taxi_driver/"),
)
```

```python
env = TaxiDriver(params).env
```

```python
algorithm = BruteForce(
    env=env,
    params=params,
    policy=Random()
)
```

```python
algorithm.run()
```

```python

TaxiDriverPlots.plot_brute_force_plot(
    policy=algorithm.computed_policy,
    algorithm=algorithm,
    env=env,
    params=params,
)
```
