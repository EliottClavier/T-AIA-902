Résultats avec theta à 0,5 :

- Le nombre moyen de pas est de 13 pour atteindre l'objectif
- Le temps "d'entrainement" est de 6.6 secondes

```python
from pathlib import Path

from common.params import TaxiDriverParams
from common.algorithms import ValueIteration
from common.policies import Max
from common.environments import TaxiDriver
from common.plots import TaxiDriverPlots
import time
```

```python
params = TaxiDriverParams(
    n_episodes=10000,
    n_runs=100,
    learning_rate=0.85,
    gamma=0.99,
    epsilon=1.0,
    min_epsilon=0.001,
    manual_decay_rate=0.001,
    random_seed=True,
    seed=123,
    max_n_steps=100,
    theta=0.5,  # Convergence threshold for Value Iteration
    savefig_folder=Path("./static/img/taxi_driver/"),
    savemodel_folder=Path("./static/models/taxi_driver/"),
)
```

```python
env = TaxiDriver(params).env
```

```python

algorithm = ValueIteration(
        env=env,
        params=params,
        policy=Max() 
    )
```

```python

start_time = time.time()
algorithm.run()

print(f"Execution time: {time.time() - start_time:.2f} seconds")
```

```python
TaxiDriverPlots.plot(
    policy=algorithm.computed_policy,
    algorithm=algorithm,
    env=env,
    params=params,
)
```
